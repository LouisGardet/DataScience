# 1) Chargement et manipulation du jeu de données

Le format csv est un format assez simple, et il est facilement exploitable via la librairie pandas. On peut commencer à charger les données et à les regarder.

````python
import pandas as pd
import numpy as np

tickets = pd.read_csv('titanic_ticket_info.csv')

print("Informations sur les tickets")
print("Les cinq premières lignes du jeu de données")
print(tickets.head(5))
print("Un résumé statistique des informations numériques du fichier")
print(tickets.describe())
print("D'autres informations sur le jeu de données des tickets")
print(tickets.info())

````

## Exercice 1.a : analyse des valeurs vides

On a pu voir plus tôt que certaines colonnes ont des valeurs vides ou nulles, on cherche maintenant à savoir quel pourcentage de valeurs sont nulles, et les colonnes affectées. Pour faire cela, il faut d'abord avoir la liste des valeurs vides, puis compter le nombre de valeurs vides par colonne, puis diviser par le nombre total pour avoir le pourcentage.

On peut trouver les valeurs nulles via la méthode isnull() (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.isnull.html), et faire savoir combien de valeurs sont nulles en faisant la somme via la méthode sum() (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.sum.html).

````python
#À vous de compléter cet encadré
null = tickets.isnull()
count = null.sum()
end = count / len(tickets) * 100
print(end)

print((tickets.isnull().sum() / len(tickets)) * 100)
````

## Exercice 1.b : chargement des données de passagers

En s'inspirant du chargement de l'autre fichier, à vous de charger et d'examiner les données des passagers. Celles-ci se trouvent dans le fichier titanic_passenger_info.csv

````python
#À vous de compléter cet encadré
passengers = pd.read_csv('titanic_passenger_info.csv')
````

## Exercice 1.c : ajout d'informations plus ou moins pertinentes

On cherche à ajouter des informations à notre jeu de données. Toutefois, nous ne sommes pas sûrs de savoir quelles informations sont utiles, alors nous décidons que la première lettre des champs de l'intitulé du ticket et du nom du passager sont sûrement des informations pertinentes. À vous de rajouter ces colonnes. Pour rappel, on peut rajouter une nouvelle colonne via le code :

```df['nouvelle_col'] = df.apply(lambda x : fonction(x), axis=1)```

````python
#À vous de compléter cet encadré
passengers['new_col'] = passengers.apply(lambda x : x['Name'][0], axis=1)
tickets['new_col'] = tickets.apply(lambda x : x['Ticket'][0], axis=1)
print(passengers)
print(tickets)
````

## Exercice 1.d : jointure des deux tables

On veut maintenant n'avoir qu'une seule table avec toutes les informations. À vous de joindre les deux tables.

````python
#À vous de compléter cet encadré
merge_df = pd.merge(tickets, passengers, on='PassengerId')
merge_df
````

## Exercice 1.e : normalisation de données numériques

On trouve que le prix n'est pas facilement lisible, et qu'il le serait beaucoup plus s'il c'était une valeur entre 0 et 1. À vous de faire une nouvelle colonne avec le prix qui ne contient que des valeurs entre 0 et 1.

````python
#À vous de compléter cet encadré
merge_df['price'] = merge_df['Fare'] / 1000
merge_df
````

## Exercice 1.f : statistiques descriptives

On veut maintenant s'intéresser un peu plus au jeu de données, et voir quel est le pourcentage de femme et d'hommes sur le bateau.

````python
#À vous de compléter cet encadré
femme = merge_df[merge_df['Sex'] == 'female']
homme = merge_df[merge_df['Sex'] == 'male']
print(len(femme) / len(merge_df) * 100)
print(len(homme) / len(merge_df) * 100)
````

## Exercice 2.a : survie en fonction du sexe

On veut représenter la survie des passagers en fonction du sexe, une première idée et d'afficher cette statistique grâce à trois camemberts : on commence par un pour montrer la répartition par sexe.

Pour voir comment faire des camemberts avec plotly, voici la documentation : https://plotly.com/python/pie-charts/

````python
#À vous de compléter cet encadré
import plotly.express as px

# dans passenger, je veux faire sortir le pourcentage de décès en fonction du sexe et du passengerId
data = [len('femme'), (len('merge_df') - len('femme'))]
labels = ["Femme", "Homme"]
fig = px.pie(values = data, names= labels)
fig.show()
merge_df
````

### Exercice 2.a.2 : survie des hommes

Pour représenter la survie des passagers en fonction du sexe, nous devons maintenant faire un deuxième camembert pour montrer la survie pour les hommes.Et pour finir il faut maintenant un camembert pour montrer la survie pour les femmes.

````python
#À vous de compléter cet encadré
homme_survive = homme[homme['Survived']== 1]
data_h = [len(homme) - len(homme_survive), len(homme_survive)]
labels_h = ['Décédé', 'Vivant']
fig = px.pie(values = data_h, names = labels_h)
fig.show()
````

### Exercice 2.a.3 : survie des femmes

Et pour finir il faut maintenant un camembert pour montrer la survie pour les femmes.

````python
#À vous de compléter cet encadré
femme_survive = femme[femme['Survived']== 1]
data_f = [len(femme) - len(femme_survive), len(femme_survive)]
labels_f = ['Décédée', 'Vivante']
fig = px.pie(values = data_f, names = labels_f)
fig.show()
````

### Exercice 2.a.4 : représentation générale

Ce genre de représentation n'est pas optimal, alors on décide plutôt d'afficher ces infos dans un histogramme, dont on peut lire la documentation ici : https://plotly.com/python/bar-charts/

````python
#À vous de compléter cet encadré
import plotly.graph_objects as go

fig = go.Figure(data=[
    go.Bar(name='Homme', x=data_h, y=[20, 14, 23]),
    go.Bar(name='Femmme', x=femme, y=[12, 18, 29])
])
# Change the bar mode
fig.update_layout(barmode='group')
fig.show()
````
